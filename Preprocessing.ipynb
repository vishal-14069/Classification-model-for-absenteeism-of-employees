```python
import pandas as pd 
import numpy as np
```

```python
data = pd.read_csv("Absenteeism_data.csv")
data
```

```python
data.info()
```

```python
data_copy= data.copy()
#data_copy
```

```python
data_copy= data_copy.drop(['ID'],axis=1) # dropping the age and the ID columns because they are label variables,they wont help much for the analysis
data_copy
```

```python
#data
```

```python
# Creating one hot encoding for the Reasons for Absence and the Age columns
dummies_absence= pd.get_dummies(data_copy['Reason for Absence'])
#dummies_absence
```

```python
# Checking if the number one is present in each row, if sum of all the values in the row is 1 then it implies that each row has a reason for absence
pd.options.display.max_columns=None
dummies_absence['Check']= dummies_absence.sum(axis=1)
dummies_absence
```

```python
dummies_absence['Check'].sum(axis=0)
```

```python
dummies_absence= dummies_absence.drop(['Check'],axis=1)
dummies_absence= dummies_absence.drop([0],axis=1) # dropping 0 column to avoid multi-colliniearity 
#dummies_absence
```

```python
# Grouping the reasons of absence to avoid too many columns in the dataset

group1= dummies_absence.loc[:,1:14].max(axis=1)
group2= dummies_absence.loc[:,15:17].max(axis=1)
group3= dummies_absence.loc[:,18:21].max(axis=1)
group4= dummies_absence.loc[:,22:].max(axis=1)
```

```python
data_copy= pd.concat([data_copy, group1,group2,group3,group4],axis=1)
data_copy= data_copy.drop(['Reason for Absence'],axis=1)
data_copy
```

```python
data_copy.columns
```

```python
data_copy_columns= ['Date','Transportation Expense','Distance to Work','Age','Daily Work Load Average','Body Mass Index','Education',                  'Children','Pets', 'Absenteeism Time in Hours',
                    'group 1', 'group 2','group 3','group 4']
data_copy.columns= data_copy_columns
data_copy
data_copy.columns
```

```python
data_copy_reorder= ['group 1', 'group 2', 'group 3','group 4','Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education', 'Children',
       'Pets', 'Absenteeism Time in Hours']
data_copy= data_copy[data_copy_reorder]
data_copy
```

```python
data_copy['Date']= pd.to_datetime(data_copy['Date'])
data_copy['Date']= pd.to_datetime(data_copy['Date'],format='%d/%m/%Y')
data_copy['Date']
```

```python
# Extracting the month value from the date column in the dataset

months=[]
for i in range(700):
    months.append(data_copy['Date'][i].month)
    
data_copy['Month']= months
#data_copy
```

```python
# Extracting the day of the week

def date_to_weekday(date_value):
    return date_value.weekday()

data_copy['Day of the week']= data_copy['Date'].apply(date_to_weekday)
data_copy
```

```python
data_copy= data_copy.drop(['Date'],axis=1)
data_copy
```

```python
data_copy.columns
```

```python
data_reorder2= ['group 1', 'group 2', 'group 3', 'group 4','Month','Day of the week', 'Transportation Expense',
       'Distance to Work', 'Age', 'Daily Work Load Average', 'Body Mass Index',
       'Education', 'Children', 'Pets', 'Absenteeism Time in Hours']
data_copy= data_copy[data_reorder2]
data_copy
```

```python
# In the education column, 1 is considered as high school, 2 as graduate, 3 as post-graduate,4 as masters
# Mapping 1 as 0 and 2,3,4 as 1 so that we will know the people who went to high school as 0 and people who went to college as 1
data_copy['Education']= data_copy['Education'].map({1:0,2:1,3:1,4:1})
data_copy
```

```python
data_copy['Education'].value_counts()
```

```python
data_copy.to_csv('preprocessed_absenteeism_data.csv', index=False)
```
